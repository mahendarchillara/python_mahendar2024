**Outlier_Analysis**

```python
#non_outlier

age_data=churndf["age"]
q1=np.percentile(churndf["age"],25)
q2=np.percentile(churndf["age"],50)
q3=np.percentile(churndf["age"],75)
IQR=q3-q1
lb=q1-1.5*IQR
ub=q3+1.5*IQR

print(lb,ub)

con1=age_data>lb
con2=age_data<ub
con3=con1&con2
non_outlier_df=churndf[con3]

non_outlier_df[["age"]]
```

```python
age_data=churndf["age"]
q1=np.percentile(churndf["age"],25)
q2=np.percentile(churndf["age"],50)
q3=np.percentile(churndf["age"],75)
IQR=q3-q1
lb=q1-1.5*IQR
ub=q3+1.5*IQR
con1=age_data<lb
con2=age_data>ub
con4=con1|con2

outlier_df=churndf[con4]

outlier_df[["age"]]
```

#np.where

```python
churn_data=pd.read_csv("Churn.csv")

churn_data["gender"].replace("nan",np.nan,inplace=True)
mode_gender=churn_data["gender"].mode()
churn_data["gender"].fillna(mode_gender[0],inplace=True)

churn_data["maximum_days_inactive"].replace("nan",np.nan,inplace=True)
mode_maximum_days_inactive=churn_data["maximum_days_inactive"].mode()
churn_data["maximum_days_inactive"].fillna(mode_maximum_days_inactive[0],inplace=True)

churn_data["churn"].replace("nan",np.nan,inplace=True)
mode_churn=churn_data["churn"].mode()
churn_data["churn"].fillna(mode_churn[0],inplace=True)
```

```python
churn_data['maximum_days_inactive'].astype('int')
churn_data['churn'].astype('int')
```

```python
con1=churn_data["gender"]=="Male"
churn_data["gender"]=np.where(con1,0,1)
churn_data

con2=churn_data["multi_screen"]=="yes"
churn_data["multi_screen"]=np.where(con2,0,1)

con3=churn_data["mail_subscribed"]=="yes"
churn_data["mail_subscribed"]=np.where(con3,0,1)

churn_data
```

#3LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
churn_data=pd.read_csv("Churn.csv")

churn_data["gender"].replace("nan",np.nan,inplace=True)
mode_gender=churn_data["gender"].mode()
churn_data["gender"].fillna(mode_gender[0],inplace=True)

churn_data["maximum_days_inactive"].replace("nan",np.nan,inplace=True)
mode_maximum_days_inactive=churn_data["maximum_days_inactive"].mode()
churn_data["maximum_days_inactive"].fillna(mode_maximum_days_inactive[0],inplace=True)

churn_data["churn"].replace("nan",np.nan,inplace=True)
mode_churn=churn_data["churn"].mode()
churn_data["churn"].fillna(mode_churn[0],inplace=True)

churn_data["maximum_days_inactive"].astype('int')
churn_data['churn'].astype('int')

le=LabelEncoder()
for i in cat_columns:
    churn_data[i]=le.fit_transform(churn_data[i])
churn_data
```
